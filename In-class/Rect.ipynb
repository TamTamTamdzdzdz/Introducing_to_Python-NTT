# Question 2:
Write a Python class named **Rect** constructed by a length and width and two methods that will compute the area and the perimeter of a rectangle.

```python
class Rect:
    def __init__(self,a,b) -> None:
        self.width=a
        self.height=b
    def __str__(self) -> str:
        return '<Width = '+str(self.width)+', Height = '+str(self.height)+'>'
    def area(self):
        return self.width*self.height
    def perimeter(self):
        return 2*(self.width+self.height)

a = Rect(4, 5)
print(a)
print(a.area(), a.perimeter())
```
