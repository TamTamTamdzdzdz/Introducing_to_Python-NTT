# Question 1:
Write a class Fraction with the following specifications:

\- Data Attributes: **num**, **deno** (two integers)

\- Methods: 

* **\_\_init\_\_**, **\_\_str\_\_**
* **add**, **subtract** two fractions
* **invert** the fraction
* **\_\_float\_\_** convert a fraction to a float

```python
class Fraction:
    def __init__(self,num,deno) -> None:
        self.num=num
        self.deno=deno
    def __str__(self) -> str:
        return str(self.num)+'/'+str(self.deno)
    def __float__(self):
        return self.num/self.deno
    def inverse(self):
        return Fraction(self.deno,self.num)
    def __add__(self,other):
        newnum=self.num*other.deno+self.deno*other.num
        newdeno=self.deno*other.deno
        return Fraction(newnum,newdeno)
    def __sub__(self,other):
        newnum=newnum=self.num*other.deno-self.deno+other.num
        newdeno=self.deno*other.deno
        return Fraction(newnum,newdeno)

f1 = Fraction(3, 5)
f2 = Fraction(6, 7)
print(f1)
print(f1 + f2)
print(f1.inverse())
print(float(f1))
```
