# Question 1:
Write an abstract class namely **Figure** that contains two following abstract methods: **getArea()** and **getPerimeter()**

Then, write three subclasses that inherit from the Figure class and implement the two above abstract methods:

* **Triangle**: attributes a, b, c representing the side lengths of a triangle
* **Rectangle**: attributes h, w representing the height and the width of a rectangle
* **Square**: attribute a representing the side length of a square

```python
import math
from abc import ABC, abstractclassmethod
class Figure(ABC):
    def getArea(self):
        pass
    def getPerimeter(self):
        pass
class Triangle(Figure):
    def __init__(self,a,b,c) -> None:
        super().__init__()
        self.a=a
        self.b=b
        self.c=c
    def getPerimeter(self):
        return self.a+self.b+self.c
    def getArea(self):
        temp=self.getPerimeter()/2
        return math.sqrt(temp*(temp-self.a)*(temp-self.b)*(temp-self.c))

class Rectangle(Figure):
    def __init__(self,a,b) -> None:
        super().__init__()
        self.a=a
        self.b=b
    def getArea(self):
        return self.a*self.b
    def getPerimeter(self):
        return 2*(self.a+self.b)
    

class Square(Figure):
    def __init__(self,a) -> None:
        super().__init__()
        self.a=a
    def getArea(self):
        return self.a**2
    def getPerimeter(self):
        return 4*self.a

obj1 = Triangle(3, 4, 5)
obj2 = Rectangle(4, 5)
obj3 = Square(5)
print("%.2f %.2f %.2f" % (obj1.getArea(), obj2.getArea(), obj3.getArea()))
print("%.2f %.2f %.2f" % (obj1.getPerimeter(), obj2.getPerimeter(), obj3.getPerimeter()))
print(isinstance(obj1, Figure), isinstance(obj2, Figure), isinstance(obj3, Figure))
```
